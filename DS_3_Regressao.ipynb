# Importando as bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,  mean_absolute_error, mean_absolute_percentage_error
```

# Carregando o conjunto de dados de diabetes


```python
diabetes = load_diabetes()
X_reg = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y_reg = pd.DataFrame(diabetes.target, columns=['target'])
```

```python
diabetes.feature_names
```

```python
diabetes.data
```

```python
display(X_reg)
display(y_reg)
```

```python
df_diabetes = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df_diabetes['target'] = pd.Series(diabetes.target)

df_diabetes.hist(figsize=(12,8))
```

```python
df_diabetes.corr()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_diabetes.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlação entre Variáveis Numéricas")
plt.show()
```

# Dividindo os dados em conjunto de treinamento e teste para regressão


```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(df_diabetes[diabetes.feature_names], df_diabetes['target'], test_size=0.2, random_state=42)
```

# Treinando o modelo de regressão linear

```python
reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)
y_pred_reg = reg_model.predict(X_test_reg)
```

# Avaliando o modelo de regressão linear


```python
mse = mean_squared_error(y_test_reg, y_pred_reg)
print("Erro quadrático médio (MSE) para regressão:", mse)

mae = mean_absolute_error(y_test_reg, y_pred_reg)
print("Erro médio absoluto (MAE) para regressão:", mae)
3
mape = mean_absolute_percentage_error(y_test_reg, y_pred_reg)
print("Erro médio percentual absoluto (MAPE) para regressão:", mape)
```

# Plotando a linha de regressão


```python

plt.plot(range(0,len(y_pred_reg)), y_test_reg, color='black', linewidth=3)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg, color='blue', linewidth=3)
plt.show()
```

```python
print(reg_model.coef_.reshape(reg_model.coef_.shape[0]),type(reg_model.coef_))
```

**Testando o Knn Regressor**

```python
from sklearn.neighbors import KNeighborsRegressor


reg_model_knn = KNeighborsRegressor(n_neighbors=5)
reg_model_knn.fit(X_train_reg, y_train_reg)
y_pred_reg_knn = reg_model_knn.predict(X_test_reg)
```

```python
mse = mean_squared_error(y_test_reg, y_pred_reg_knn)
print("Erro quadrático médio (MSE) para regressão:", mse)

mae = mean_absolute_error(y_test_reg, y_pred_reg_knn)
print("Erro médio absoluto (MAE) para regressão:", mae)
3
mape = mean_absolute_percentage_error(y_test_reg, y_pred_reg_knn)
print("Erro médio percentual absoluto (MAPE) para regressão:", mape)
```

```python
plt.plot(range(0,len(y_pred_reg)), y_test_reg, color='black', linewidth=3)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg, color='red', linewidth=2)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg_knn, color='blue', linewidth=2)
plt.show()
```

**Testando a DecisionTree**

```python
from sklearn.tree import DecisionTreeRegressor

reg_model_dt = DecisionTreeRegressor(max_depth=50)
reg_model_dt.fit(X_train_reg, y_train_reg)
y_pred_reg_dt = reg_model_dt.predict(X_test_reg)
```

```python
mse = mean_squared_error(y_test_reg, y_pred_reg_dt)
print("Erro quadrático médio (MSE) para regressão:", mse)

mae = mean_absolute_error(y_test_reg, y_pred_reg_dt)
print("Erro médio absoluto (MAE) para regressão:", mae)
3
mape = mean_absolute_percentage_error(y_test_reg, y_pred_reg_dt)
print("Erro médio percentual absoluto (MAPE) para regressão:", mape)
```

```python
plt.plot(range(0,len(y_pred_reg)), y_test_reg, color='black', linewidth=3)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg, color='red', linewidth=2)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg_knn, color='blue', linewidth=2)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg_dt, color='cyan', linewidth=2)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

reg_model_rf = RandomForestRegressor(n_estimators=1000)#n_estimators=10
reg_model_rf.fit(X_train_reg, y_train_reg)
y_pred_reg_rf = reg_model_rf.predict(X_test_reg)
```

```python
mse = mean_squared_error(y_test_reg, y_pred_reg_rf)
print("Erro quadrático médio (MSE) para regressão:", mse)

mae = mean_absolute_error(y_test_reg, y_pred_reg_rf)
print("Erro médio absoluto (MAE) para regressão:", mae)
3
mape = mean_absolute_percentage_error(y_test_reg, y_pred_reg_rf)
print("Erro médio percentual absoluto (MAPE) para regressão:", mape)
```

```python
plt.plot(range(0,len(y_pred_reg)), y_test_reg, color='black', linewidth=3)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg, color='red', linewidth=2)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg_knn, color='blue', linewidth=2)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg_dt, color='cyan', linewidth=2)
plt.plot(range(0,len(y_pred_reg)), y_pred_reg_rf, color='green', linewidth=2)
plt.show()
```
